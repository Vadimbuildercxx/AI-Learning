```python
import pandas as pd
```

```python
data = pd.read_csv('../input/adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data["sex"].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
data[data["sex"] == "Female"]["age"].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
import numpy as np
data.groupby(["native-country"])["native-country"].apply(lambda native_country: native_country.value_counts()/ len(data.index))
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
data[(data["salary"] == "<=50K")]["age"].agg([np.mean, np.std])
```

```python
data[(data["salary"] == ">50K")]["age"].agg([np.mean, np.std])
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python

df_salary_more_50k = pd.DataFrame(data[(data["salary"] == ">50K")]["education"])
is_hight_schools = ((df_salary_more_50k == "Bachelors") | (df_salary_more_50k == "Prof-school") |
                   (df_salary_more_50k == "Assoc-acdm") | (df_salary_more_50k == "Assoc-voc") |
                   (df_salary_more_50k == "Masters") | (df_salary_more_50k == "Doctorate"))
df_salary_more_50k.insert(loc=1, column="high school education", value=is_hight_schools)
df_salary_more_50k.head(10)
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
data.groupby(["race", "sex"])["race", "sex", "age"].agg([np.mean, np.std, np.min, np.max])
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
mens = pd.DataFrame(data[data["sex"] == "Male"])
married_str = ["Married-AF-spouse", "Married-civ-spouse", "Married-spouse-absent"]
single_str = ["Divorced", "Never-married","Separated"]
is_hight_schools = (data["marital-status"].isin(married_str))

mens.insert(loc=1, column="is married", value=is_hight_schools)

pd.crosstab(mens["is married"], mens["salary"])
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
print(data["hours-per-week"].max())
quantity_of_99hours_workers = data[data["hours-per-week"] == 99]["hours-per-week"].value_counts()
print(data[(data["hours-per-week"] == 99) & (data["salary"] == ">50K")]["hours-per-week"].value_counts() / quantity_of_99hours_workers)
```
